## 1C i.

Time domain features that are usually used in time-series classification include the minimum, maximum, mean, median, standard deviation, first quartile, third quartile, range (max - min), interquartile range (Q3 - Q1), variance (how far each value is from the mean), and kurtosis (how often outliers occur in a distribution).

## 1C ii.

```python
#extraction function
def features_extraction(data, timeseries_names):
    features = {}
    for k, col in enumerate(timeseries_names, 1): #loop through each column name and index from 1
        features[f'min{k}'] = data[col].min()
        features[f'max{k}'] = data[col].max()
        features[f'mean{k}'] = data[col].mean()
        features[f'median{k}'] = data[col].median()
        features[f'std{k}'] = data[col].std()
        features[f'1st_quart{k}'] = data[col].quantile(0.25)
        features[f'3rd_quart{k}'] = data[col].quantile(0.75)
    return features

#get features from each dataframe
features_list = [features_extraction(df, df.columns) for df in combined_dataframes]
df_timedom_features = pd.DataFrame(features_list)

df_timedom_features
```

## 1C iii.

```python
def bootstrap_CI(data, n_bootstrap=1000, ci=90):
    boot_stds = np.zeros(n_bootstrap) #store boostrapped means

    num_samples = len(data)
    
    for i in range(n_bootstrap): #creating boostrapped samples to find std
        boot_samples = np.random.choice(data, size=num_samples, replace=True)
        boot_stds[i] = np.std(boot_samples)
    
    l_bound = np.percentile(boot_stds, (100 - ci) / 2) #lower CI bound
    u_bound = np.percentile(boot_stds, 100 - (100 - ci) / 2) #upper CI bound
    
    return np.std(data), l_bound, u_bound
```

```python
results_dict = {}
for col in df_timedom_features.columns: #loop through time domain features
    std = bootstrap_CI(df_timedom_features[col])
    lower = bootstrap_CI(df_timedom_features[col])
    upper = bootstrap_CI(df_timedom_features[col])
    results_dict[col] = {'Standard Deviation': std, 'Lower Bound Confidence Interval': lower, 'Upper Bound Confidence Interval': upper}

    
results_df = pd.DataFrame(results_dict)
results_df
```

## 1C iv.

The 3 most important domain features are mean, median, and standard deviation. The mean shows the behavior of the data because it is affected by outliers. The standard deviation shows how far away the values are from the mean, and lastly the median shows how skewed the data can be by considering the middle value.
